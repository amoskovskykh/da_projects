# Survey monkey dataset analysis 
## processing data from wide to long format + basic aggregations

```python
import pandas as pd
import os
```

```python
path = os.getcwd()
```

```python
df = pd.read_excel(path + "//edited_data.xlsx", sheet_name = "Edited_Data")
```

```python
df
```

```python
df.columns
```

### Dropping empty columns or those which are irrelevant for the further analysis

```python
columns_to_drop = ['Start Date', 'End Date', 'Email Address', 'First Name', 'Last Name', 'Custom Data 1']
```

```python
df = df.drop(columns=columns_to_drop)
```

```python
df.columns
```

### Transforming data from wide to long format

```python
id_vars = list(df.columns)[:8]
unpiv_vars = list(df.columns)[8:]
```

```python
df_melted = df.melt(id_vars=id_vars, value_vars=unpiv_vars, var_name="Question_subquery", value_name="Answer")
df_melted.head()
```

### Right now I'm operating with question_subquestion column, but client just wants the original question column
So I will load another sheet which contains the raw question list

```python
question_df = pd.read_excel(path + "//edited_data.xlsx",sheet_name = "Question")
```

```python
question_df.drop(columns = ['Raw Question', 'Subquestion Raw' ], inplace=True)
```

```python
question_df.columns
```

### Joining datasets on the "Question_subquery", "Question+Subquestion" columns

```python
df_merged = pd.merge(df_melted, question_df, how="inner", left_on="Question_subquery", right_on="Question+Subquestion")
```

```python
df_merged
```

```python
# just checking the lengths
print("Original Data lenght: ", len(df_melted))
print("Merged Data lenght: ", len(df_merged))
```

```python
#dropping rows where Answer is NaN
respondent = df_merged[df_merged["Answer"].notna()]
respondent
```

```python
# calculating the number of unique respondents to a question and sorting in desc order by Respondent ID
respondent_data = respondent.groupby("Question")["Respondent ID"].nunique().reset_index().sort_values(by = 'Respondent ID', ascending = False)
respondent_data
```

```python
# renaming the column
respondent_data.rename(columns={"Respondent ID":"Number of Unique Respondents"}, inplace = True)
# or respondent_data.columns = ["Question", "Numnber of Unique Respondents"]
```

```python
respondent_data
```

```python
# appening new data to the main dataframe
df_merged_2 = pd.merge(df_merged, respondent_data, how="inner", left_on="Question", right_on="Question")
```

```python
df_merged_2
```

### Calculating the number of the same answer given to a (question, subquestion) combination

```python
# now I group by question+subquestion combination and answer
same_answer = df_merged_2.groupby(["Question+Subquestion","Answer"])["Respondent ID"].nunique().reset_index()
same_answer.rename(columns={"Respondent ID":"Same Answer"}, inplace=True)
same_answer
```

```python
df_merged_3 = pd.merge(df_merged_2, same_answer, how="left", on=["Question+Subquestion", "Answer"])
df_merged_3["Same Answer"] = df_merged_3["Same Answer"].fillna(0)
```

```python
df_merged_3
```

### Renaming columns for a final touch and specifying column order

```python
id_vars[1:]
```

```python
df_final = df_merged_3.rename(columns={'Identify which division you work in. - Response':"Division Primary",
 'Identify which division you work in. - Other (please specify)':"Division Secondary",
 'Which of the following best describes your position level? - Response': "Position",
 'Which generation are you apart of? - Response':"Generation",
 'Please select the gender in which you identify. - Response':"Gender",
 'Which duration range best aligns with your tenure at your company? - Response':"Duration range expected",
 'Which of the following best describes your employment type? - Response':"Empoyment type"})
```

```python
df_final.columns
```

```python
df_final.drop(columns=["Question_subquery"], inplace = True)
```

```python
df_final
```

```python
new_order = ['Respondent ID', 'Division Primary', 'Division Secondary', 'Position',
       'Generation', 'Gender', 'Duration range expected', 'Empoyment type', 'Question+Subquestion','Question', 'Answer',
       'Number of Unique Respondents', 'Same Answer']
```

```python
df_final = df_final.reindex(columns = new_order)
```
